**Analysis for the Missing Values**

The below code outputs all the rows where there is any missing values.

* For the column *"normalized-losses"* total missing values is 41, which is 20% of the data. This should not Imputed (fill the missing) with any value as it will simply add biasness to the whole dataset, so **Drop this column**
* For 4 rows, the *"price"* column is not having any values.These 4 rows cannot be used for any prediction, so  **delete these 4 rows**
* Wherever *"bore"* is NaN, value for *"stroke"* is also missing. So **drop these 4 rows**
* Similarly *"horsepower"* and *"peak-rpm"* columns have 2 records with missing values. **Drop these 2 rows**
* Finally *"num-of-doors"* column has 2 missing values. **Drop these 2 rows**

```python
missing_col_names = automobile_df.columns[automobile_df.isna().any()].tolist()
missing_col_names
```

```python
# Find out all the rows with missing values (NaN). We could have also used 
# "automobile_df[automobile_df.isnull().any(axis=1)]" but since we have many more missing
# values in the 'normalized-losses' column we wanted to selectively select the column name.

automobile_df[automobile_df['num-of-doors'].isnull() | 
              automobile_df['bore'].isnull()         |
              automobile_df['stroke'].isnull()       |
              automobile_df['horsepower'].isnull()   |
              automobile_df['peak-rpm'].isnull()     | 
              automobile_df['price'].isnull()
             ]
```

Based on the analysis above, treat the missing values onto the dataframe. 
1. Drop the "normalized-losses" column
2. Drop all rows where missing value is found

```python
#First drop the "normalized-losses" column
automobile_df.drop('normalized-losses', axis=1, inplace=True)

#Remove all the rows where NaN values are found. 12 rows should be removed
null_cols = ['num-of-doors', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
automobile_df.dropna(axis=0, inplace=True, subset = null_cols)

postmissing_df_shape = automobile_df.shape
print(postmissing_df_shape)
```

```python
automobile_df.info()
```

Create a Function to find out all the unique values for dataframe columns. It returns a dictionary which can be anaylzed to see if  the datatype of the columns can be changed to **"category"**

```python
def get_unique_vals(df):
    dict_un={}
    for i in df.columns:
        if df[i].dtype == 'O':
            dict_un[i]= df[i].unique().tolist()
    return dict_un

print (automobile_df['symboling'].unique().tolist())

dict_un = get_unique_vals(automobile_df)
dict_un
```

All the columns with datatype as 'Object' have categorical values. Also, the *'symboling'* column is a categorical. The *'make'* column could be considered as a categorical value, but I thought not to do it.

So we will convert the datatype for all these columns as **'Categorical'**

```python
cat_columns = ['fuel-type', 
               'aspiration', 
               'num-of-doors', 
               'body-style', 
               'drive-wheels', 
               'engine-location', 
               'engine-type', 
               'num-of-cylinders', 
               'fuel-system',
               'symboling'
              ]

for col in cat_columns:
    automobile_df[col] = automobile_df[col].astype('category')

automobile_df.info()
```

### Summary of the Data Cleaning activity ###

First loaded the data into a DataFrame automobile_df. **The dataset had a shape of 205 rows, 26 columns.**

Initial analysis suggested that there are missing values in the text file. Instead of keeping it blank the missing values are marked with the character "?". So while reading the text file into the DataFrame, considered "?" as missing value.

The column *"normalized-losses"* has 20% of data which are missing. So instead of imputing the column with mean/median value it was best to remove the column itself for any analysis. Filling missing values to this column might introduce biasness to the data.

The columns *[num-of-doors, bore, stroke, horsepower, peak-rpm, price]* have few missing values and based on the earlier analysis we have decided to remove the rows where we have missing values in any of these columns. 

**After removing all observations including missing values, 193 rows and 25 columns are retained.**

Also analyzed the unique values in the columns which were imported as "object" datatype to validate if any of them can be converted into "Categorical" values. Based on the analysis the columns *[fuel-type, aspiration, num-of-doors, body-style, drive-wheels, engine-location, engine-type, num-of-cylinders, fuel-system, symboling]* were converted into **Categorical**.
